```python
columns = ['범죄대분류', '범죄중분류', '서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종',
       '경기도', '강원도', '충청북도', '충청남도', '전라북도', '전라남도', '경상북도', '경상남도', '제주도']
```

```python
df = df[columns].copy()
```

```python
df_strong = df[df["범죄대분류"] == "강력범죄"]
df_strong = df_strong.drop("범죄대분류", axis=1).copy()
df_strong = df_strong.set_index("범죄중분류")
df_strong = df_strong.T
df_strong
```

## 컬러 설정

```python
colors = sns.color_palette("Pastel2")
```

```python
plt.figure(figsize=(15, 10)) 
ax = sns.heatmap(df_strong, annot=True, fmt="d", cmap='RdPu',
                 linewidths=2, annot_kws={"size": 17})

plt.title("지역별 강력범죄 범죄별 그래프", fontsize=20)
```

##지역별 강력범죄 횟수 그래프

```python
df_strong2 = df_strong.sum(axis=1)
df_strong2
```

```python
ax = df_strong2.plot.bar(figsize=(17, 12), color=colors, edgecolor='black', linewidth=3, rot=0, fontsize=15)

for p in ax.patches: 
    ax.annotate("%.0f 회" % p.get_height(), (p.get_x() + p.get_width()/2., p.get_height() - 30), 
       ha='center', va='center', fontsize=15, color='black', xytext=(0, 20), 
       textcoords='offset points')
plt.box(False)
plt.title("지역별 강력범죄 종합 횟수",fontsize=20)
```
