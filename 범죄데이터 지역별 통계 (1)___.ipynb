## 지역별 강력범죄 그래프

```python
df_strong = df[df["범죄대분류"] == "강력범죄"]
df_strong = df_strong.drop("범죄대분류", axis=1).copy()
df_strong = df_strong.set_index("범죄중분류")
df_strong = df_strong.T
df_strong
```

### 지역별 강력범죄 heatmap 그래프

```python
plt.figure(figsize=(17, 12)) 
res = sns.heatmap(df_strong, annot=True, fmt="d", cmap='RdPu',
                 linewidths=2, annot_kws={"size": 17})

res.set_xticklabels(res.get_xmajorticklabels(), fontsize = 13)
res.set_yticklabels(res.get_ymajorticklabels(), fontsize = 13)
res.set_xlabel("")

plt.title("지역별 강력범죄별 그래프", fontsize=20)

#서울-강제추행, 경기도-강제추행이 가장 많은 횟수를 보여줌
```

### 지역별 강력범죄 횟수 그래프

```python
df_strong2 = df_strong.sum(axis=1)
df_strong2

#열을 기준으로 모두 더하기
```

```python
ax = df_strong2.plot.bar(figsize=(17, 12), color=colors, edgecolor='black', linewidth=3, rot=0, fontsize=14)

#그래프에 숫자 추가
for p in ax.patches: 
    ax.annotate("%.0f 회" % p.get_height(), (p.get_x() + p.get_width()/2., p.get_height() - 30), 
       ha='center', va='center', fontsize=15, color='black', xytext=(0, 20), 
       textcoords='offset points')

#그래프의 박스를 안보이게끔 설정
plt.box(False)
plt.title("지역별 강력범죄 종합 횟수",fontsize=20)
```
